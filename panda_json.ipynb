```python
import pandas as pn
import numpy as np

from requests import get
from gzip import decompress
```

```python
res = pn.read_json(
        decompress(
            get("https://github.com/sameergarg/scala-elasticsearch/blob/master/conf/recipeitems-latest.json.gz?raw=true").content),
        lines=True
        )
```

```python
res
```

##### find the name of recipe that has the longest ingredient list

```python
res.iloc[res['ingredients'].str.split('\n').str.len().sort_values(ascending=False).head(1).index[0]]['name']
```

##### how many of the recipes are for breakfast food

```python
res['description'].str.contains('breakfast', regex=True).sum()
```

##### How many of the recipes contain precisely 3 ingredients

```python
(res['ingredients'].str.split('\n').str.len() == 3).sum()
```
